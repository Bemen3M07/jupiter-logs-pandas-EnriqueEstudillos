# Codigo replicado de prueba

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame basado en la tabla proporcionada
data = {
    "Profesor": ["Emili", "Isabel", "Jose", "Emili", "Isabel"],
    "Alumno": ["Xavier", "Juan", "Carlos", "Maradona", "Paquito"],
    "M01": [8, 6, 5, 5, 5],
    "M02": [6, 4, 8, 4, 9],
    "M03": [7, 5, 8, 3, 9],
    "M04": [9, 7, 7, 8, 8],
    "M05": [4, 5, 7, 8, 4],
}

notas = pd.DataFrame(data)
print("DataFrame")
print(notas)
print("               ")

promedio = notas[["M01", "M02", "M03", "M04", "M05"]].mean()
print("Promedio por modulo Axis 0")
print(promedio)
print("               ")

promedio = notas[["M01", "M02", "M03", "M04", "M05"]].mean(axis = 1)
print("Promedio por Alumno Axis 1")
print(promedio)
print("               ")

promedio = notas.groupby('Alumno', as_index=False)[["M01", "M02", "M03", "M04", "M05"]].mean()
print("Promedio por alumno por materia")
print(promedio)
print("               ")

# Grafico de las notas de M01 Y M02 de los alumnos
print("Notas M01 y M02 por alumno")
notas.plot(x="Alumno", y=["M01", "M02"])
plt.rcParams["figure.figsize"] = (10,5)
plt.show()

#Grafico de las notas de M02 de cada alumno
print("Notas M02 por alumno")
x_values = notas["Alumno"]
y_values = notas["M02"]
plt.plot(x_values, y_values, color="red")
plt.legend(["M02"])
plt.show()
```

**Implementar un log en un programa es crucial para monitorizar, analizar y depurar el comportamiento del código durante su ejecución.**

**Ejemplo 1:**
En un sistema de comercio electrónico, un cliente no puede completar una compra porque el sistema devuelve un error 500. Mediante un log, se puede ver que el error se debe a una conexión fallida con la base de datos, lo que permite al desarrollador solucionarlo directamente.

**Ejemplo 2:**
Un servicio de streaming registra que el tiempo de carga de un vídeo se ha incrementado significativamente. El log muestra que una consulta en la API está tardando demasiado, lo que apunta a un problema de rendimiento a optimizar.

**Ejemplo 3:**
En un sistema bancari, un log mostra que un usuari va fer un accés no autoritzat a una funció restringida. Això permet detectar comportaments maliciosos i prendre mesures preventives o correctives.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame basado en la tabla proporcionada
data = {
    "Profesor": ["Emili", "Isabel", "Jose", "Emili", "Isabel"],
    "Alumno": ["Xavier", "Juan", "Carlos", "Maradona", "Paquito"],
    "M01": [8, 6, 5, 5, 5],
    "M02": [6, 4, 8, 4, 9],
    "M03": [7, 5, 8, 3, 9],
    "M04": [9, 7, 7, 8, 8],
    "M05": [4, 5, 7, 8, 4],
}

notas = pd.DataFrame(data)
print("DataFrame")
print(notas)
print("               ")

promedio = notas[["M01", "M02", "M03", "M04", "M05"]].mean()
print("Promedio por modulo")
print(promedio)
print("               ")

promedio = notas.groupby('Alumno', as_index=False)[["M01", "M02", "M03", "M04", "M05"]].mean()
print("Promedio por alumno por materia")
print(promedio)
print("               ")

# Grafico de las notas de M01 Y M02 de los alumnos
print("Notas M01 y M02 por alumno")
notas.plot(x="Alumno", y=["M01", "M02"])
plt.rcParams["figure.figsize"] = (10,5)
plt.show()
```
